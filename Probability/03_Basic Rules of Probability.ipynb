## <b>Basic Rules of Probability</b>
### <b>1. Important Definitions</b>
In probability, understanding the basic definitions is crucial for grasping more complex concepts.

- <b>Experiment:</b> An action or process that leads to one or more outcomes (e.g., rolling a die). Deterministic experiments have predictable outcomes, while probabilistic or random experiments have uncertain outcomes.

- <b>Outcome:</b> A possible result of an experiment (e.g., rolling a 3 on a die).

- <b>Probability (P):</b> A measure of the likelihood that an event will occur, ranging from 0 (impossible) to 1 (certain).

- <b>P(AUB):</b> The probability of event A or event B occurring.

- <b>P(A∩B):</b> The probability of both events A and B occurring.

- <B>Sample Space (S):</b> The set of all possible outcomes of an experiment.

- <b>Event (E):</b> A subset of the sample space.

- <b>Mutually Exclusive Events:</b> Two events that cannot occur at the same time (e.g., rolling a 2 and rolling a 5 on a die). P(A∩B)=0

- <b>Independent Events:</b> Two events where the occurrence of one does not affect the other. P(A∩B)=P(A)⋅P(B)

- <b>Certain Event:</b> An event that is guaranteed to happen, with a probability of 1.

- <b>Impossible Event</b>: An event that cannot happen, with a probability of 0.

- <b>Exhaustive Events:</b> A set of events that cover the entire sample space, meaning at least one of them must occur.

### <b>Example:</b>
Tossing a die:

Sample space: S = {1, 2, 3, 4, 5, 6}

Event (E): Rolling an even number = {2, 4, 6}

### <b>2. The Complement Rule</b>
The complement of an event A is the event that A does not occur.

- Notation: Aᶜ
- Rule: P(Aᶜ) = 1 − P(A)
#### <b>Example:</b>
If the probability of rain today is 0.3, the probability it won’t rain is:

P(No Rain) = 1 − 0.3 = 0.7

## <b>3. The Addition Rule</b>
Used to calculate the probability of the union of two events.

#### <b>For general events:</b>
P(A or B) = P(A) + P(B) − P(A and B)

#### <b>For mutually exclusive events:</b>
P(A or B) = P(A) + P(B)

#### <b>Example:</b>
P(A) = 0.4, P(B) = 0.5, P(A and B) = 0.2
P(A or B) = 0.4 + 0.5 − 0.2 = 0.7

### <b>4. The Multiplication Rule</b>
Used to find the probability that two events occur together.

#### <bFor Independent events:</b>
P(A and B) = P(A) × P(B)

Example: A event A is rolling a 3 on a first throw of a die, and event B is rolling an even number on second throw.

- P(A) = 1/6 (rolling a 3)
- P(B) = 3/6 (rolling a 2, 4, or 6)
- P(A and B) = P(A) × P(B) = (1/6) × (3/6) = 1/12
#### <b>For Dependent events:</b>
P(A and B) = P(A) × P(B | A)

Example: A bag contains 5 red and 3 blue balls. Two balls are drawn without replacement. Let:

- Event A be drawing a red ball first.

- Event B be drawing a red ball second.

- P(A) = 5/8 (5 red balls out of 8)

- After removing one red ball, there are 4 red left out of 7 total: P(B/A) = 4/7

- P(A and B) = (5/8) X (4/7)= 20/56 = 5/14

So the probability of drawing two red balls <b>without replacement</b> is 5/14

### <b>5. Independent vs Dependent Events - More examples</b>
- <b>Independent:</b> The outcome of one event does not affect the other.
    - Example: Tossing two coins.
- <b>Dependent:</b> One event affects the probability of the other.
    - Example: Drawing two cards without replacement.

### <b>6. Mutually Exclusive Events</b>
Two events are <b>mutually exclusive</b> if they cannot happen at the same time.

<b>Example:</b>
- Event A: Rolling a 2
- Event B: Rolling a 5
- These are mutually exclusive because a die can't show both at once.
