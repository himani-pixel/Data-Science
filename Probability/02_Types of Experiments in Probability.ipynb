## <b>Types of Experiments in Probability</b>
### <b>Types of Experiments</b>
<b>1. Deterministic</b> – Always produces the same outcome when repeated under identical conditions.
<b> 2. Probabilistic / Random</b> – Outcome is uncertain and may vary even if the experiment is repeated under identical conditions.

### <b>Sample Space</b>
The sample space is the set of all possible outcomes of a probabilistic experiment.

#### Examples:

- <b>Rolling a Die</b> → { 1, 2, 3, 4, 5, 6 }
- <b>Tossing 2 Coins</b> → { (HH), (HT), (TH), (TT) }

#### <b>For Coins</b>
- Number of possible outcomes = <b>2ⁿ</b><br>
(where n is the number of coins)

#### <b>For Dice</b>
- Number of possible outcomes = <b>6ⁿ</b><br>
(where n is the number of dice)

### <b>For Cards</b>
A standard deck contains <b>52 cards:</b>
- <b>26 Red Cards</b>

  - 13 ♥️ Hearts → A, 2–10, J, Q, K
  - 13 ♦️ Diamonds → A, 2–10, J, Q, K

- <b>26 Black Cards</b>

  - 13 ♠️ Spades → A, 2–10, J, Q, K
  - 13 ♣️ Clubs → A, 2–10, J, Q, K

### <b>Conclusion</b>
Understanding the basic types of experiments and the concept of sample space is essential for solving probability problems. Knowing the number of possible outcomes for coins, dice, and cards helps in calculating probabilities accurately. Mastery of these fundamentals forms the foundation for more advanced topics in probability and statistics.
