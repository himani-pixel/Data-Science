## <b>Introduction to Probability distributions</b>
### <b>What is a Probability Distribution?</b>
Imagine you're analyzing real-world data — sales, weather, clicks, heights, etc. You’ll notice:

- Some values are more <b>likely</b> than others.
- There's a <b>pattern</b> to how data is spread.
That pattern is described by a <b>probability distribution.</b>

#### <b>In short:</b>
```
"A probability distribution tells you how likely different outcomes are."
```

### <b>What is a Probability Distribution?</b>
A <b>probability distribution</b> is a <b>mathematical function</b> or <b>table</b> that:

- Assigns <b>probabilities</b> to all possible outcomes of a random process.
There are two types:

### <b>1. Discrete Probability Distribution</b>
- For outcomes you can <b>count</b> (e.g. number of heads in 3 coin tosses).
- Example: <b>Binomial Distribution</b>
```
Think: “What’s the probability I get exactly 2 heads in 3 coin tosses?”
```

### <b>2. Continuous Probability Distribution</b>
- For outcomes that can be any number in a range (e.g. height, weight).
- Example: Normal Distribution
```
Think: “What’s the probability someone’s height is between 165 cm and 170 cm?”
```

### <b>A Simple Analogy</b>
#### <b>Let’s say you roll a die:</b>
Outcomes = {1, 2, 3, 4, 5, 6}
Each has a probability of `1/6`
This is a <b>uniform distribution</b> (discrete).

Now imagine measuring people's <b>heights:</b>

- You don’t get fixed values.
- Instead, you get a <b>curve</b> — most people around average height, fewer very short or very tall.

That’s a <b>normal distribution</b> (continuous).

### <b>Why Are Distributions Useful in Data Science?</b>
1. <b>Model real-world randomness</b> (user behavior, errors, arrivals, etc.)
2. <b>Make predictions</b> (how likely is a customer to buy?)
3. <b>Run simulations</b>
