### Querying with .query()
The .query() method in pandas lets you filter DataFrame rows using a string expression — it's a more readable and often more concise alternative to using boolean indexing.

This is a cleaner, SQL-like way to filter:

```python
df.query("`BoxOffice(INR Crore)` > 200 and Genre == 'Comedy'")
```

### Here are the main rules and tips for using .query() in pandas:

1. Column names become variables <br>
You can reference column names directly in the query string:<br>

df.query("age > 25 and city == 'Delhi'")

2. String values must be in quotes<br>
Use single or double quotes around strings in the expression:<br>

df.query("name == 'Harry'")

If you have quotes inside quotes, mix them:

df.query('city == "Mumbai"')

3. Use backticks for column names with spaces or special characters<br>
If a column name has spaces, use backticks (`):<br>

df.query("`first name` == 'Alice'")

4. You can use @ to reference Python variables<br>
To pass external variables into .query():<br>

age_limit = 30
df.query("age > @age_limit")<br>

5. Logical operators<br>
Use these:<br>

and, or, not — instead of &, |, ~ , ==, !=, <, >, <=, >=<br>

Bad:

df.query("age > 30 & city == 'Delhi'")  # ❌

Good:

df.query("age > 30 and city == 'Delhi'")  # ✅

6. Chained comparisons<br>
Just like Python:<br>

df.query("25 < age <= 40")

7. Avoid using reserved keywords as column names<br>
If you have a column named class, lambda, etc., you’ll need to use backticks:<br>

df.query("`class` == 'Physics'")

8. Case-sensitive<br>
Column names and string values are case-sensitive:<br>

df.query("City == 'delhi'")  # ❌ if actual value is 'Delhi'

9. .query() returns a copy, not a view<br><br>
The result is a new DataFrame. Changes won't affect the original unless reassigned:

filtered = df.query("age < 50")
