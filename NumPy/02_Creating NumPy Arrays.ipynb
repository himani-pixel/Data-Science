## NumPy Array

```python
import numpy as np
arr1 = np.array([2 , 4 , 6 , 8 , 10])
arr2 = np.array([[3 , 6 , 9] , [12 , 15 , 18]])
print("1D - Array:",arr1)
print("2D - Array:\n",arr2)
```

### Array only with ZEROS

```python
np.zeros((3,3))
```

### Array only with ONES

```python
np.ones((3,3))
```

### Array filled with Specified nos

```python
np.full((3 , 3),4)
```

### Identity Matrix

```python
np.eye((2))
```

### Range of Array

```python
np.arange(1 , 10, 2)
```

### Evenly Spaced Array

```python
np.linspace(0 , 1 , 5)
```

- <b>NumPy offers powerful shortcutts to create arrays without loops</b>

# Properties of NumPy Array

```python
arr = np.array([2 , 4 , 6 , 8 , 10])
print("The shape of the arr is:",arr.shape)
print("The dimension of the array is:",arr.ndim)
print("The size of the Array is:",arr.size)
print("The DataType of the array is:",arr.dtype)
```

- <b> Numpy are strongly typed i.e all elements have the same data type.</b>

## Changing DType

```python
arr = np.array([1, 2, 3], dtype=np.float32)  # Explicit type
print(arr.dtype)  # float32

arr_int = arr.astype(np.int32)  # Convert float to int
print(arr_int)  # [1 2 3]
```
