Here are some common NumPy methods that are frequently used for statistical
and mathematical operations:

#### 1. np.mean() â€“ Compute the mean (average) of an array.

```python

import numpy as np

# Creating sample arrays
arr = np.array([2, 4, 6, 7, 9, 10])
arr1 = np.array([2, 4, 6, 8, 10])
arr2 = np.array([3, 6, 9, 12, 15])

print("Array 1:", arr)
print("Array 2:", arr1)
print("Array 3:", arr2)
```

```python

# Mean - Average of all elements
print("Mean of arr:", np.mean(arr))
```

```python

# Standard Deviation - Measure of spread of values
print("Standard Deviation of arr:", np.std(arr))
```

```python

# Variance - Square of standard deviation
print("Variance of arr:", np.var(arr))
```

```python

# Minimum and Maximum value in array
print("Minimum value in arr:", np.min(arr))
print("Maximum value in arr:", np.max(arr))
```

```python

# Sum of all elements in array
print("Sum of arr:", np.sum(arr))
```

```python

# Square root of each element
print("Square root of each element in arr:", np.sqrt(arr))
```

```python

# Exponential of each element
print("Exponential of each element in arr:", np.exp(arr))
```

```python

# Logarithmic values (base e)
print("Natural log (ln) of arr1:", np.log(arr1))
```

```python

# Trigonometric functions
angles = np.array([0, np.pi/6, np.pi/4, np.pi/3, np.pi/2])
print("Sine values:", np.sin(angles))
print("Cosine values:", np.cos(angles))
print("Tangent values:", np.tan(angles))
```

```python

# Rounding functions
arr3 = np.array([1.2, 2.5, 3.7, 4.4, 5.9])
print("Original:", arr3)
print("Rounded:", np.round(arr3))
print("Floor:", np.floor(arr3))
print("Ceil:", np.ceil(arr3))
```
