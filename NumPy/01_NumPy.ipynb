## NumPy: Numerical Python

### NumPy's Superpowers:
- Faster than Python lists (C-optimized backend)
- Uses less memory (efficient storage)
- Supports vectorized operations (no explicit loops needed)
- Has built-in mathematical functions

### SpeedTest of NumPy array and Python List 

```python
# Using Python List 
import time
```

```python
size = 1_000_000
l1 = list(range(size))
l2 = list(range(size))
start  =  time.time()

result = [x + y for x , y in zip(l1 , l2)]
end = time.time()
print(end - start )
```

```python
#Using Numpy Array
import numpy as np
import time 
size = 1_000_000
l1 = np.array(list(range(size)))
l2 = np.array(list(range(size)))
start  =  time.time()

result = l1 + l2
end = time.time()
print(end - start )
```
