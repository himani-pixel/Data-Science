## <b>Introduction to Flask</b>
### <b>Why Flask?</b>
- Lightweight Python web framework
- Ideal for serving models and exposing APIs

### <b>Installation</b>

```python
pip install flask
```

```python
from flask import Flask

app = Flask(__name__)

@app.route('/')
def hello_world():
    return '<h2>Hello, Data Scientist!</h2>'


app.run(debug=True)
```
